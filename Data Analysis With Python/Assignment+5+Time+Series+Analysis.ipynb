## Problem 1
You have been tasked with analyzing the market data of your favorite stock. Write a Python program to complete the following tasks:
+ Create a DataFrame named `my_stock` by loading the daily prices (High, Low, Open, Close, Adjusted Close) and volumes of your chosen stock from the first trading day of 2023 to today. Display `my_stock`.

+ Calculate the simple daily return and log daily return of the stock. Add the results as two new columns to `my_stock` with the name of **'Daily_simple_return'** and **'Daily_log_return'**, respectively.

+ Calculate the daily cumulative return of the stock and add the results as a new column named **'Daily_cum_return'** to `my_stock`.

+ Create a line plot to visualize the trend of the stock's **daily cumulative return** over time. Properly label the x-axis, y-axis, and title of the plot.


```python
###
### YOUR CODE HERE
###
import yfinance as yf
my_stock=yf.download("WRBY","2023-1-1")
my_stock.columns = [c[0] for c in my_stock.columns]
display(my_stock)
```

```python
import numpy as np
my_stock_close=my_stock["Close"]
my_stock["Daily_simple_return"]=my_stock_close.pct_change()
my_stock["Daily_log_return"]=np.log(my_stock_close).diff()
my_stock
```

```python
my_stock["Daily_cum_return"] = np.cumprod(my_stock["Daily_log_return"]+1)-1
my_stock
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 8))
my_stock["Daily_cum_return"].plot(color="blue")
plt.title("WRBY Cumulative Return")
plt.xlabel("Date") 
plt.ylabel("Cumulative Return") 
plt.show()
```

## Problem 3
Expand the analysis you have done in the previous problems by completing the following tasks:
+ Calculate the 20-day moving averages and standard deviations of the stock's closing prices. Add the results as two new columns to `my_stock` with the name of **'SMA_20'** and **'STD_20'**, respectively.

+ Calculate the 20-day Bollinger Bands. Add the results as two new columns to `my_stock` with the name of **'Upper_BB'** and '**Lower_BB'**, respectively.

+ Create a plot to visualize the 20-day moving average, upper and lower Bollinger Bands. Fill the area between the upper and lower bands with a color and transparency you prefer. Properly label the x-axis, y-axis, legend, and title of the plot.


```python
###
### YOUR CODE HERE
###
my_stock["SMA_20"] = my_stock_close.rolling(20).mean() 
my_stock["STD_20"] = my_stock_close.rolling(20).std()
display(my_stock)
```

```python
my_stock["Upper_BB"] = my_stock["SMA_20"]+2*my_stock["STD_20"]
my_stock["Lower_BB"] = my_stock["SMA_20"]-2*my_stock["STD_20"]
my_stock
```

```python
import seaborn as sns
my_stock_bollinger = my_stock[["SMA_20","Upper_BB","Lower_BB"]]
display(my_stock_bollinger)
```

```python
plt.figure(figsize = (12, 6))
sns.lineplot(my_stock_bollinger)
plt.fill_between(x=my_stock_bollinger.index,
                y1=my_stock_bollinger["Lower_BB"] ,
                y2=my_stock_bollinger["Upper_BB"] , 
                color="blue", 
                alpha=0.25 
                )
plt.ylabel("$") 
plt.title("WRBY 20-Day Bollinger Bands") 
plt.show()
```

## Problem 5

You have been tasked with forecasting the future price of your chosen stock. Write a Python program to complete the following tasks:
+ From the `my_stock` DataFrame you created in Problem 1, create two new DataFrames named `train` and `test`, respectively. The `train` DataFrame includes the closing price for all the trading days in 2023 and 2024. The `test` DataFrame includes the closing price for the year-to-date in 2025.

+ Modify `train` and `test` so they both have two columns named **'ds'** and **'y'**. The **'ds'** column shows the dates and the **'y'** column shows the stock closing price on those dates.

+ Create a prophet model that accounts for weekly, yearly, and U.S. holiday seasonalities.

+ Train the prophet model using the `train` DataFrame.
+ Use the trained model to forecast the stock price in the first 60 days in 2025. Display the forcasting result.
+ Create a plot that visualizes the actual and predicted prices. The plot should display the model's confidence interval.
+ Visualize the different components of the forecast.


```python
###
### YOUR CODE HERE
###
train = my_stock.loc["2023":"2024","Close"]
test = my_stock.loc["2025","Close"]
train = train.reset_index().rename(columns={'Date': 'ds', 'Close': 'y'})
test = test.reset_index().rename(columns={'Date': 'ds', 'Close': 'y'})
display(test)
display(train)
```

```python
from prophet import Prophet
prophet_model = Prophet(yearly_seasonality=True,weekly_seasonality=True,daily_seasonality=True)
prophet_model.add_country_holidays(country_name='US')
prophet_model.fit(train)
import warnings
warnings.filterwarnings('ignore')
future = prophet_model.make_future_dataframe(periods=60, 
                     freq="D")
forecast = prophet_model.predict(future)
display(forecast)
```

```python
prophet_model.plot(forecast)
plt.show()
```

```python
prophet_model.plot_components(forecast)
plt.show()
```

## Problem 6
You have been tasked to evaluate the performance of the Prophet model you trained in the Problem 5. Write a Python program to compelete the following tasks:
+ Create a new DataFrame by merging the stock's actual and predicted prices (including the predicted upper and lower boundaries) on all the dates included in the `test` DataFrame. Set the **'ds'** column as the row index of the new DataFrame.

+ Create a plot with 4 lines to visualize the stock's true price, predicted price, predicted upper boundary, and predicted lower boundary. Fill the area between the predicted upper and lower boundaries with a color and transparency you prefer.

```python
###
### YOUR CODE HERE
###
import pandas as pd
forecast_price = forecast[['ds', 'yhat','yhat_lower', 'yhat_upper']]
test_forecast = pd.merge(test,forecast_price,on="ds",how="inner")
test_forecast = test_forecast.set_index("ds")
display(test_forecast)
```

```python
test_forecast.plot(figsize = (12, 8))
plt.fill_between(x = test_forecast.index, 
                 y1 = test_forecast['yhat_lower'], 
                 y2 = test_forecast['yhat_upper'], 
                 color = 'blue',
                 alpha = .25)
plt.title('Forecasted vs Actual Prices')
plt.show()
```
