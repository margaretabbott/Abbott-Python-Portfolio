## Problem 4

Expand the analysis you have done in the previous problem by completing the following tasks:
+ Output a summary of the rows and columns in the `wsoc` DataFrame.
+ The **'Height'** column shows the players' heights in the **Feet'Inch** string fromat (i.e. 5'6). Convert them to the interger version of inches (i.e. 66).
+ Find the tallest Goalkeeper and output a message stating her name and height.

```python
###
### YOUR CODE HERE
###
wsoc.info()
```

```python
height_to_inches = lambda h:int(h.split("'")[0])*12+int(h.split("'")[1])
wsoc['Converted Height']=wsoc['Height'].apply(height_to_inches)
wsoc.head()
```

```python
wsoc_goalies=wsoc[wsoc['Position']=='Goalkeeper']
tallest_goalie=wsoc_goalies.loc[wsoc_goalies['Converted Height'].idxmax()]
print(f"The tallest goalie on the womens soccer roster is {tallest_goalie.name} who is {tallest_goalie['Height']}")
```

## Problem 5
You have been tasked with analyzing the Titanic passengers. Write a Python program to complete the following tasks:
+ Create a DataFrame named `titanic` by importing the data from the CSV file located at https://www3.nd.edu/~xduan/data-sets/titanic.csv.
+ Display the last 3 rows of `titanic`.
+ From `titanic`, create a new DataFrame named `survived_female` that includes the female survivors and display it.
+ Calculate the percentage of female surviors and print a message stating your result.

**Note**: The percentage of female survivors should be calculated by dividing the number of female survivors by the total number of female passengers and multiplying 100.

```python
###
### YOUR CODE HERE
###
titanic=pd.read_csv('https://www3.nd.edu/~xduan/data-sets/titanic.csv')
titanic.tail(3)
```

```python
survived=titanic['Survived']==1
female=titanic['Sex']=='female'
survived_female=titanic[(survived)&(female)]
display(survived_female)
```

```python
total_females=titanic[female]
percentage_survived_female=round((len(survived_female)/len(total_females))*100,2)
print(f'Roughly {percentage_survived_female}% of female passengers survived the sinking of the Titanic')
```
